```
import time
import random
import matplotlib.pyplot as plt

# number of elements in the list
min,max = 0,5000



```

# **Compare O(n) vs O(n^2) vs O(n^3)**

```
def o_n(arr):
  for i in range(1, len(arr)):
    if (i < i+1):
      name = 'teste'

def o_n2(arr):
  for i in range(1, len(arr)):
    for j in range(1, len(arr)):      
      if (i < i+1):
        name = 'teste'


def o_n3(arr):

  if (len(arr) > 10 ):
    for i in range(1, len(arr)):
      for j in range(1, len(arr)):
        for k in range(1, len(arr)):
          if (i < i+1):
            name = 'teste'



i_min,i_max,step = 1,30,1

cost_on = list()
cost_on2 = list()
cost_on3 = list()
n_list = list()
for i in range(i_min,i_max,step):
  n_list.append(i)
  #Generate n random numbers between min and max
  randomlist = random.sample(range(min, max), i)
  
  t1 = time.clock()
  o_n(randomlist)
  t2 = time.clock()
  timePassed = (t2-t1)
  cost_on.append(timePassed)

  t1 = time.clock()
  o_n2(randomlist)
  t2 = time.clock()
  timePassed = (t2-t1)
  cost_on2.append(timePassed)

  t1 = time.clock()
  o_n3(randomlist)
  t2 = time.clock()
  timePassed = (t2-t1)
  cost_on3.append(timePassed)

plt.plot(n_list, cost_on, linewidth=3, color='red', label='O(n)')
plt.plot(n_list, cost_on2, linewidth=3, color='blue', label='O(n^2)')
plt.plot(n_list, cost_on3, linewidth=3, color='black', label='O(n^3)')
plt.legend(loc='upper right')
plt.title('O(n) functions value', fontsize=19)
plt.xlabel('n value')
plt.ylabel('Cost function')
plt.show()
```

# **Fibonacci: Iterative x Recursive**

```
def fib_iterative (n): 
    a, b = 0, 1
    for i in range(0, n):
        a, b = b, a + b
    return a


def fib_recursive(n):
    if n < 2:
        return n
    return fib_recursive(n-2) + fib_recursive(n-1)


max, step = 20,2
n_list = list()
sort_list,rec_list = list(),list()

for i in range(1,max,step):
  n_list.append(i)
  
  t1 = time.clock()
  
  fib_iterative(i)
  t2 = time.clock()  
  timePassed = (t2-t1)  
  sort_list.append(timePassed)

  t1 = time.clock()
  fib_recursive(i)
  t2 = time.clock()
  timePassed = (t2-t1)
  rec_list.append(timePassed)

plt.plot(n_list, sort_list, linewidth=3, color='red', label='Iterative')
plt.plot(n_list, rec_list, linewidth=3, color='blue', label='Recursive')
plt.legend(loc='upper right')
plt.title('Fibonacci: Iterative x Recursive', fontsize=19)
plt.xlabel('n value')
plt.ylabel('Time spend')
plt.show()
```
